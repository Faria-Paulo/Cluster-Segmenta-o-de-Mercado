## Segmentação de Clientes com base no Comportamento de Uso do Cartão de Crédito.

O conjunto de dados para este notebook consiste no comportamento de uso do cartão de crédito de clientes, com 18 características comportamentais. A segmentação dos clientes pode ser utilizada para definir estratégias de marketing.


**Conteúdo deste documento:**

* Pré-processamento de Dados
* Clusterização usando KMeans
* Interpretação dos Clusters
* Visualização dos Clusters usando PCA


```python
# Importando as bilbiotecas necessárias

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Coneção com a tabela
data= pd.read_csv("/content/drive/MyDrive/csv/CC GENERAL.csv")
print(data.shape)
data.head()
```

##Procesamento de dados

#### Estatísticas Descritivas dos Dados

```python
data.describe()
```

####Lidando com dados faltantes

```python
data.isnull().sum().sort_values(ascending=False).head()
```

Imputando valores faltantes com mean.

```python
data.loc[(data['MINIMUM_PAYMENTS'].isnull()==True),'MINIMUM_PAYMENTS']=data['MINIMUM_PAYMENTS'].mean()
data.loc[(data['CREDIT_LIMIT'].isnull()==True),'CREDIT_LIMIT']=data['CREDIT_LIMIT'].mean()
```

```python
data.isnull().sum().sort_values(ascending=False).head()
```

####Lidando com os Outliers

Ao remover os outliers, podemos perder muitas linhas, pois há muitos outliers no conjunto de dados. Portanto, estamos criando faixas/ ranges para lidar com valores extremos.



```python
columns=['BALANCE', 'PURCHASES', 'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'CREDIT_LIMIT',
        'PAYMENTS', 'MINIMUM_PAYMENTS']

for c in columns:

    Range=c+'_RANGE'
    data[Range]=0
    data.loc[((data[c]>0)&(data[c]<=500)),Range]=1
    data.loc[((data[c]>500)&(data[c]<=1000)),Range]=2
    data.loc[((data[c]>1000)&(data[c]<=3000)),Range]=3
    data.loc[((data[c]>3000)&(data[c]<=5000)),Range]=4
    data.loc[((data[c]>5000)&(data[c]<=10000)),Range]=5
    data.loc[((data[c]>10000)),Range]=6
```

```python
columns=['BALANCE_FREQUENCY', 'PURCHASES_FREQUENCY', 'ONEOFF_PURCHASES_FREQUENCY', 'PURCHASES_INSTALLMENTS_FREQUENCY',
         'CASH_ADVANCE_FREQUENCY', 'PRC_FULL_PAYMENT']

for c in columns:

    Range=c+'_RANGE'
    data[Range]=0
    data.loc[((data[c]>0)&(data[c]<=0.1)),Range]=1
    data.loc[((data[c]>0.1)&(data[c]<=0.2)),Range]=2
    data.loc[((data[c]>0.2)&(data[c]<=0.3)),Range]=3
    data.loc[((data[c]>0.3)&(data[c]<=0.4)),Range]=4
    data.loc[((data[c]>0.4)&(data[c]<=0.5)),Range]=5
    data.loc[((data[c]>0.5)&(data[c]<=0.6)),Range]=6
    data.loc[((data[c]>0.6)&(data[c]<=0.7)),Range]=7
    data.loc[((data[c]>0.7)&(data[c]<=0.8)),Range]=8
    data.loc[((data[c]>0.8)&(data[c]<=0.9)),Range]=9
    data.loc[((data[c]>0.9)&(data[c]<=1.0)),Range]=10
```

```python
columns=['PURCHASES_TRX', 'CASH_ADVANCE_TRX']

for c in columns:

    Range=c+'_RANGE'
    data[Range]=0
    data.loc[((data[c]>0)&(data[c]<=5)),Range]=1
    data.loc[((data[c]>5)&(data[c]<=10)),Range]=2
    data.loc[((data[c]>10)&(data[c]<=15)),Range]=3
    data.loc[((data[c]>15)&(data[c]<=20)),Range]=4
    data.loc[((data[c]>20)&(data[c]<=30)),Range]=5
    data.loc[((data[c]>30)&(data[c]<=50)),Range]=6
    data.loc[((data[c]>50)&(data[c]<=100)),Range]=7
    data.loc[((data[c]>100)),Range]=8
```

```python
data.drop(['CUST_ID', 'BALANCE', 'BALANCE_FREQUENCY', 'PURCHASES',
       'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE',
       'PURCHASES_FREQUENCY',  'ONEOFF_PURCHASES_FREQUENCY',
       'PURCHASES_INSTALLMENTS_FREQUENCY', 'CASH_ADVANCE_FREQUENCY',
       'CASH_ADVANCE_TRX', 'PURCHASES_TRX', 'CREDIT_LIMIT', 'PAYMENTS',
       'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT' ], axis=1, inplace=True)

X= np.asarray(data)
```

Normalizando valores Inputados

```python
scale = StandardScaler()
X = scale.fit_transform(X)
X.shape
```

## MODELANDO

#### Clusterizando usando Kmeans

```python
n_clusters=30
cost=[]
for i in range(1,n_clusters):
    kmean= KMeans(i)
    kmean.fit(X)
    cost.append(kmean.inertia_)
```

```python
plt.plot(range(1, n_clusters), cost, marker='o')
plt.xlabel('Número de clusters')
plt.ylabel('Inércia')
plt.title('Método do Cotovelo (Elbow Method)')
plt.show()
```

#### 6 será o número de clusters


```python
kmean= KMeans(6)
kmean.fit(X)
labels=kmean.labels_
```

```python
clusters=pd.concat([data, pd.DataFrame({'cluster':labels})], axis=1)
clusters.head()
```

## Interpretação dos Clusters

```python
for c in clusters:
    grid= sns.FacetGrid(clusters, col='cluster')
    grid.map(plt.hist, c)
```

* **Cluster0** Pessoas com limite de crédito médio a alto que fazem todos os tipos de compras

* **Cluster1** Este grupo tem mais pessoas com pagamentos em atraso que sacam dinheiro adiantado com mais frequência

* **Cluster2** Pessoas que gastam pouco, com limites de crédito médios a altos, que fazem compras principalmente a prazo

* **Cluster3** Pessoas com limite de crédito alto que sacam mais dinheiro adiantado

* **Cluster4** Grandes gastadores com limite de crédito alto que fazem compras caras

* **Cluster5** Pessoas que não gastam muito dinheiro e que têm limite de crédito médio a altocredit limit

(O número dos clusters podem mudar ao rodar o código novamente)

## Visualização dos Clusters

#### Usando PCA para transformar os dados para 2 dimensões para visualização


```python
dist = 1 - cosine_similarity(X)

pca = PCA(2)
pca.fit(dist)
X_PCA = pca.transform(dist)
X_PCA.shape
```

```python
x, y = X_PCA[:, 0], X_PCA[:, 1]

colors = {0: 'red',
          1: 'blue',
          2: 'green',
          3: 'yellow',
          4: 'orange',
          5:'purple'}

names = {0: 'Que fazem todo tipo de compra', #Que fazem todo tipo de compra
         1: 'pessoas que devem pagamentos', #pessoas que devem pagamentos
         2: 'quem compra parcelado', # quem compra parcelado
         3: 'que aceitam mais dinheiro adiantado',# que aceitam mais dinheiro adiantado
         4: 'que mais fazem compras caras',#que mais fazem compras caras
         5:'que não gastam muito dinheiro'} #que não gastam muito dinheiro

df = pd.DataFrame({'x': x, 'y':y, 'label':labels})
groups = df.groupby('label')

fig, ax = plt.subplots(figsize=(20, 13))

for name, group in groups:
    ax.plot(group.x, group.y, marker='o', linestyle='', ms=5,
            color=colors[name],label=names[name], mec='none')
    ax.set_aspect('auto')
    ax.tick_params(axis='x',which='both',bottom='off',top='off',labelbottom='off')
    ax.tick_params(axis= 'y',which='both',left='off',top='off',labelleft='off')

ax.legend()
ax.set_title("Segmentação de Clientes com base no comportamento de uso de seus Cartões de Crédito.")
plt.show()
```
